```python
corpus = """This is a dataset for binary sentiment classification containing substantially more data than previous benchmark datasets. 
We provide a set of 25,000 highly polar movie reviews for training, and 25,000 for testing. 
There is additional unlabeled data for use as well. Raw text and already processed bag of words formats are provided. 
See the README file contained in the release for more details."""
```

```python
print(corpus)
```

# Tokenization
Using `NLTK`

`Sentance` to `Paragraph`  using `sent_tokenize` funcrion

```python
from nltk.tokenize import sent_tokenize
```

```python
document = sent_tokenize(corpus)
```

```python
print(document)
print(type(document))
```

```python
for doc in document:
    print(doc)
```

# Tokenize 

`Sentance` to `words` 

`paragraph` to `Words`

Using `word_tokenize` function.

```python
from nltk.tokenize import word_tokenize
```

```python
word = list(word_tokenize(corpus))

word
```

```python
for w in word:
    print(w)
```

# Using `wordpunct_tokenize` function

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
w_punct = wordpunct_tokenize(corpus)
w_punct
```

# Using TreebankWordTokenizer function

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
t_w_tokenize = TreebankWordTokenizer()
```

```python
t_w_tokenize.tokenize(corpus)
```
